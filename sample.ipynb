```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Tasodifiy o'zgaruvchilar soni
np.random.seed(42)
n_samples = 20

# Qora mol bozor ma'lumotlarini yaratamiz
x = np.random.rand(n_samples, 1) * 10  # Masalan, og'irlik (kg)
y = 2.5 * x + 6 + np.random.randn(n_samples, 1) * 2  # Narx (mln.so'm)


# Bir o‘zgaruvchili regressiya
def single_variable_regression(x, y):
    model = LinearRegression()
    model.fit(x, y)
    y_pred = model.predict(x)
    print("Bir o‘zgaruvchili regressiya koeffitsienti: ", model.coef_)
    print("Intercept (y akkizmasi): ", model.intercept_)
    plt.scatter(x, y, color="blue", label="Haqiqiy ma'lumot")
    plt.plot(x, y_pred, color="red", label="Bashorat qilingan chiziq")
    plt.title("Bir o‘zgaruvchili regressiya")
    plt.xlabel("Og'irlik (kg)")
    plt.ylabel("Narx (mln.so'm)")
    plt.legend()
    plt.show()

```

```python

# Ko‘p o‘zgaruvchili regressiya (3-darajali polinom)
def multi_variable_regression(x, y, degree=3):
    poly = PolynomialFeatures(degree=degree)
    x_poly = poly.fit_transform(x)

    model = LinearRegression()
    model.fit(x_poly, y)
    y_pred = model.predict(x_poly)

    print(f"Ko‘p o‘zgaruvchili regressiya (daraja {degree}) koeffitsientlari: ", model.coef_)
    print("Intercept: ", model.intercept_)

    # Vizualizatsiya qilish
    x_sorted, y_pred_sorted = zip(*sorted(zip(x.flatten(), y_pred.flatten())))
    plt.scatter(x, y, color="blue", label="Haqiqiy ma'lumot")
    plt.plot(x_sorted, y_pred_sorted, color="green", label="Bashorat polinomi")
    plt.title(f"Ko‘p o‘zgaruvchili regressiya ({degree}-daraja)")
    plt.xlabel("Og'irlik (kg)")
    plt.ylabel("Narx (mln.so'm)")
    plt.legend()
    plt.show()
```

```python

single_variable_regression(x, y)
multi_variable_regression(x, y, degree=3)
```
